```python
import tensorflow
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
from tqdm import tqdm

print('bismillahirrahmanir-raheem')
dir= 'C:/Users/Toufique11/CVPR_11/TensorFlow_11/PetImages'
```

```python
li=['Cat', 'Dog']
list=[]
imgSize=100
for c in li:
    path=os.path.join(dir,c)
    for i in tqdm(os.listdir(path)):
        try:
            tmp=cv2.imread(os.path.join(path,i), cv2.IMREAD_GRAYSCALE)
            #plt.imshow(tmp, cmap='gray')
            tmp=cv2.resize(tmp,(imgSize,imgSize))
            indx=li.index(c)
            list.append([tmp,indx])
        except:
            pass
```

```python
import pickle
print(len(list))

X=[]
Y=[]

for i,j in list:
    X.append(i)
    Y.append(j)
    

p_out=open('X.pickle','wb')
pickle.dump(X, p_out)
p_out.close()

p_out=open('Y.pickle','wb')
pickle.dump(Y, p_out)
p_out.close()


p_in= open('X.pickle', 'rb')
XX=pickle.load(p_in)
p_in.close()

p_in= open('Y.pickle', 'rb')
YY=pickle.load(p_in)
p_in.close()
```
