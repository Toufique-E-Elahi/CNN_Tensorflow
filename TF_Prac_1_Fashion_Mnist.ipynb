```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
print('bismillahir-rahmanir-raheem')
print(tf.__version__)
```

```python
dataset=keras.datasets.fashion_mnist
(trainImages,trainLabels),(testImages, testLabels)=dataset.load_data()
imgLabels=['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
print(imgLabels)
print(len(trainImages),
trainLabels.shape,
len(trainLabels), trainImages.shape)
print(type(trainImages[0]))
print(trainImages[0].shape)
plt.figure()
plt.imshow(trainImages[0])
plt.show()

print(trainImages[100])
print(trainImages[100].shape)
```

```python
trainImages=trainImages/255.0
testImages=testImages/255.0

print(trainImages[100])
print(trainImages[100].shape)


plt.figure()
plt.imshow(trainImages[0])
plt.show()
```

```python
plt.figure()
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.imshow(trainImages[i])
plt.show()

```

```python
plt.figure(figsize=(20,20))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.imshow(trainImages[i],cmap=plt.cm.binary)
    plt.xticks([])
    plt.yticks([])
    plt.xlabel(imgLabels[trainLabels[i]])
plt.show()
```

```python
model= keras.Sequential([keras.layers.Flatten(input_shape=(28,28)),
                         keras.layers.Dense(256, activation='relu'),
                         keras.layers.Dense(10, activation='relu')
                        ])
```

```python
model.compile(optimizer='adam',
             loss=keras.losses.SparseCategoricalCrossentropy(),
             metrics=['accuracy'])
model.fit(trainImages,trainLabels,epochs=10)
```

```python
test_loss,test_accuracy=model.evaluate(testImages,testLabels)
```

```python
newModel=keras.Sequential([model,
                           keras.layers.Softmax()])
```

```python
p=newModel.predict(testImages)
print(len(p))
print(np.argmax(p[0]))
```
